This first code cell sets up the seconds timer.

```python
import time

def countdown():
  user_input = int(input("Enter the time in seconds(2500 for Pomodoro): "))
  count = user_input;
  #create a while loop
  while count:
    #use the divmod() method to calculate the time in mins and seconds and store them in their respective variable
    minutes, seconds = divmod(user_input, 60)
    print( '{:02d}:{:02d}'. format(minutes, seconds), end = '\r')
    #Let the program wait for 1 second
    time.sleep(1)
    #Decrement the while loop
    count -= 1

  #Once the countdown is done, print the below statement:
  print("The countdown is done!!")
  print("Take a " + str((user_input / 5.0) * 0.0166667) + " minute break, then run this cell again once you're done!")

 
```

This next block of code will open the webcam.

```python
from IPython.display import HTML, Audio 
from google.colab.output import eval_js 
from base64 import b64decode 
import numpy as np 
import io 
from PIL import Image 
 
VIDEO_HTML = """ 
<video autoplay 
 width=%d height=%d style='cursor: pointer;'></video> 
<script> 
 
var video = document.querySelector('video') 
 
navigator.mediaDevices.getUserMedia({ video: true }) 
  .then(stream=> video.srcObject = stream) 
   
var data = new Promise(resolve=>{ 
  video.onclick = ()=>{ 
    var canvas = document.createElement('canvas') 
    var [w,h] = [video.offsetWidth, video.offsetHeight] 
    canvas.width = w 
    canvas.height = h 
    canvas.getContext('2d') 
          .drawImage(video, 0, 0, w, h) 
    video.srcObject.getVideoTracks()[0].stop() 
    video.replaceWith(canvas) 
    resolve(canvas.toDataURL('image/jpeg', %f)) 
  } 
}) 
</script> 
""" 
def take_photo(filename='photo.jpg', quality=0.8, size=(800,600)): 
  display(HTML(VIDEO_HTML % (size[0],size[1],quality))) 
  data = eval_js("data") 
  binary = b64decode(data.split(',')[1]) 
  f = io.BytesIO(binary) 
  return np.asarray(Image.open(f)) 
#   with open(filename, 'wb') as f: 
#     f.write(binary) 
#   return len(binary) 
 
 
img = take_photo() # click 
 
 
import matplotlib.pyplot as plt 
 
plt.figure(figsize=(10,10)) 
plt.imshow(img) 
plt.show()
```
